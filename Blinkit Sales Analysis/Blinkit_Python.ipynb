## Charts

### Total Sales by Fat Content

```python
sales_fat = df.groupby("Item Fat Content")["Sales"].sum().reset_index()

plt.figure(figsize=(5,5))

plt.pie(sales_fat["Sales"],
        labels=sales_fat["Item Fat Content"],
        autopct='%1.1f%%',
        startangle=90,
        wedgeprops={'width':0.4})

plt.title("Total Sales by Fat Content")

plt.show()
```

### Total Sales by Item Type

```python
sales_itemtype = df.groupby("Item Type")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))

plt.bar(sales_itemtype.index, sales_itemtype.values)

plt.xticks(rotation=90)
plt.title("Total Sales by Item Type")
plt.xlabel("Item Type")
plt.ylabel("Sales")
plt.show()
```

### Fat Content by Outlet for Total Sales

```python
fatcontent_outlet = df.groupby(["Outlet Location Type", "Item Fat Content"])["Sales"].sum().unstack()
fatcontent_outlet = fatcontent_outlet[["Low Fat", "Regular"]]

ax = fatcontent_outlet.plot(kind='bar', figsize=(8,5), title="Fat Content by Outlet for Total Sales")
plt.xlabel("Outlet Location Tier")
plt.ylabel("Total Sales")
plt.legend(title = "Item Fat Content")
plt.tight_layout()
plt.show()
```

### Total Sales by Outlet Establishment

```python
sales_outlet = df.groupby("Outlet Establishment Year")["Sales"].sum().sort_index()

plt.figure(figsize=(9,5))
plt.plot(sales_outlet.index, sales_outlet.values, marker="o")
plt.xlabel("Outlet Establishment Year")
plt.ylabel("Sales")
plt.title("Total Sales by Outlet Establishment")
plt.tight_layout()
plt.show()
```

### Sales by Outlet Location

```python
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

plt.figure(figsize=(4,4))
plt.pie(sales_by_size, labels=sales_by_size.index, autopct='%1.1f%%', startangle=90)

plt.title('Outlet Size')
plt.tight_layout()
plt.show()
```

```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)

plt.figure(figsize=(8,3))

sns.barplot(
    x='Sales',
    y='Outlet Location Type',
    data=sales_by_location,
    order=sales_by_location['Outlet Location Type']
)

plt.title('Total Sales by Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

plt.tight_layout()
plt.show()
```
